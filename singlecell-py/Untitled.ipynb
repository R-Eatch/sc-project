```python
import os
import loompy
import scanpy as sc
import scvelo as scv
# 设置目标目录
directory = "D:/BaiduNetdiskDownload/"

# 初始化存储结果的字典
loom_dict = {}
for file in os.listdir(directory):
    if file.endswith(".loom"):
        file_path = os.path.join(directory,file)
        try:
            with loompy.connect(file_path,mode='r') as ds:
                id_str=ds.ca['CellID'][0]
                id=id_str.split(':')[0]
                filename=file.split('.')[0]
                loom_dict[filename]=id
        except Exception as e:
            print(f"Error reading {file}: {e}")
print(loom_dict)
                
```

```python
import pandas as pd

# 读取原始数据
df = pd.read_csv('./R-AG-GES23_peaks.narrowPeak', sep='\t', header=None)

# 确定第一列为非字符串的行（假设非字符串即为数字）
file_a = df[df[0].apply(lambda x: isinstance(x, int))]

# 保存为文件A
file_a.to_csv('file_A.bed', sep='\t', index=False, header=False)
```

```python
# 读取文件A
df_a = pd.read_csv('file_A.bed', sep='\t', header=None, dtype=str)

# 确保第一列是字符串
df_a[0] = df_a[0].astype(str)

# 保存修改后的文件A
df_a.to_csv('file_A_modified.bed', sep='\t', index=False, header=False)
```

```python
# 提取第一列不是非字符串的行（即剩余部分）
file_b = df[~df[0].apply(lambda x: isinstance(x, int))]

# 保存为文件B
file_b.to_csv('file_B.bed', sep='\t', index=False, header=False)
```

```python
# 重新读取修改后的文件A和文件B
df_a_modified = pd.read_csv('file_A_modified.bed', sep='\t', header=None)
df_b = pd.read_csv('file_B.bed', sep='\t', header=None)

# 合并两个DataFrame
file_c = pd.concat([df_a_modified, df_b], ignore_index=True)

# 保存为文件C
file_c.to_csv('file_C.bed', sep='\t', index=False, header=False)
```
