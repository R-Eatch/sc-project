```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
Compute cross-species mean expression for a *celltype-aware* conserved-gene list,
then build a unique Top-N `selection` dict (one gene belongs to one celltype only).

Inputs
------
1. unique_genes_long.csv         ← long-format, columns: celltype,gene,[...]
2. *_cleaned.h5ad                ← one per species, must contain layer 'normalized'
   └── AnnData.obs[GROUP_KEY]    ← cell-type annotation used as filter

Outputs
-------
1. conserved_gene_expression.csv ← long table: celltype,gene,cross_mean,M_mean,R_mean,S_mean
2. Prints + returns `selection`  ← {celltype: [Top-N genes]}

Author : <your name>
Date   : 2025-06-13
"""

from pathlib import Path
from collections import OrderedDict
import numpy as np
import pandas as pd
import scanpy as sc

# --------------------------- CONFIG ---------------------------
DATA_ROOT  = Path(r"D:/111")                   # 存放 .h5ad
DATASETS   = {"M": "M-MG", "R": "R-MG", "S": "S-MG"}   # 物种缩写→文件前缀
GENE_CSV   = Path("unique_genes_long.csv")     # 长格式保守基因集
GROUP_KEY  = "newcelltype"                     # 细胞类型字段
TOP_N      = 5
CSV_OUT    = "conserved_gene_expression.csv"
# ---------------------------------------------------------------


# ---------- 1. 读取 celltype↔gene 映射，去重保持“先到先得” ----------
def load_mapping(csv_path: Path) -> dict[str, list[str]]:
    df = pd.read_csv(csv_path)

    # ↓↓↓ 把列名统统小写，便于匹配
    df.columns = [str(c).lower() for c in df.columns]

    # 如果找不到 'celltype' 和 'gene'，就把前两列强行改名
    if not {"celltype", "gene"}.issubset(df.columns):
        if df.shape[1] < 2:
            raise ValueError("CSV needs at least two columns (celltype, gene).")
        df = df.rename(columns={df.columns[0]: "celltype",
                                df.columns[1]: "gene"})

    mapping: dict[str, list[str]] = OrderedDict()
    seen = set()
    for ct, g in zip(df["celltype"], df["gene"]):
        if pd.isna(ct) or pd.isna(g) or g in seen:
            continue
        mapping.setdefault(ct, []).append(g)
        seen.add(g)
    return mapping



# ---------- 2. 计算指定细胞类型一组基因的物种均值 ----------
def mean_expr_species(adata: sc.AnnData, genes: list[str], celltype: str) -> pd.Series:
    """
    Return Series of mean expression (index=genes) for `celltype` in one species.
    If celltype absent or none of the genes in var, Series 全 NaN.
    """
    if celltype not in adata.obs[GROUP_KEY].values:
        return pd.Series(np.nan, index=genes)

    # 取目标细胞
    idx_cells = adata.obs[GROUP_KEY] == celltype
    if idx_cells.sum() == 0:
        return pd.Series(np.nan, index=genes)

    # 限制到存在于该 AnnData 的基因，保持原顺序
    valid_genes = [g for g in genes if g in adata.var_names]
    if not valid_genes:
        return pd.Series(np.nan, index=genes)

    # 取表达矩阵并求均值
    X = adata[idx_cells, valid_genes].layers["normalized"]   # 稀疏矩阵
    means = np.asarray(X.mean(axis=0)).ravel()
    return pd.Series(means, index=valid_genes).reindex(genes)  # 保证长度一致


# ---------- 3. 构建长格式表达表 ----------
def build_expression_table(adatas: dict, mapping: dict) -> pd.DataFrame:
    recs = []
    for ct, genes in mapping.items():
        # 每物种均值
        sp_means = {sp: mean_expr_species(ad, genes, ct) for sp, ad in adatas.items()}
        stacked = pd.DataFrame(sp_means)
        stacked["cross_mean"] = stacked.mean(axis=1, skipna=True)
        # 写记录
        recs.extend(
            {
                "celltype": ct,
                "gene": g,
                "cross_mean": row["cross_mean"],
                "M_mean": row.get("M", np.nan),
                "R_mean": row.get("R", np.nan),
                "S_mean": row.get("S", np.nan),
            }
            for g, row in stacked.iterrows()
        )
    return pd.DataFrame.from_records(recs)


# ---------- 4. 生成 Top-N selection（保证唯一性） ----------
def build_selection(expr_df: pd.DataFrame, mapping: dict) -> dict:
    selection = {}
    used = set()   # 已入选基因
    for ct in mapping:            # 保持 mapping 中 celltype 的顺序
        top_genes = (
            expr_df.query("celltype == @ct and gene not in @used")
                   .sort_values("cross_mean", ascending=False)
                   .head(TOP_N)["gene"]
                   .tolist()
        )
        selection[ct] = top_genes
        used.update(top_genes)
    return selection


# ---------- 5. 主流程 ----------
def main():
    # 1) 读取映射
    mapping = load_mapping(GENE_CSV)
    print(f"[INFO] Mapping loaded: {sum(map(len, mapping.values()))} unique genes across "
          f"{len(mapping)} celltypes")

    # 2) 载入各物种 AnnData
    adatas = {}
    for sp, prefix in DATASETS.items():
        f = DATA_ROOT / f"{prefix}_cleaned.h5ad"
        ad = sc.read_h5ad(f)
        if "normalized" not in ad.layers:
            raise KeyError(f"{f} - missing `layers['normalized']`")
        adatas[sp] = ad
        print(f"[INFO] {prefix}: {ad.n_obs:,} cells, {ad.n_vars:,} genes.")

    # 3) 表达量统计
    expr_df = build_expression_table(adatas, mapping)
    expr_df.to_csv(CSV_OUT, index=False)
    print(f"[INFO] CSV saved → {CSV_OUT}  ({len(expr_df):,} rows)")

    # 4) Top-N selection
    selection = build_selection(expr_df, mapping)
    print("\n=== Top-{} 'selection' dict ===".format(TOP_N))
    for ct, genes in selection.items():
        print(f"{ct:<12} : {genes}")

    return selection


if __name__ == "__main__":
    selection = main()
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
Batch heat-map of Z-scored gene expression for multiple AnnData files,
using a celltype→gene dictionary (selection).

Author: <your name>
Date  : <today>
"""

import os
from pathlib import Path
import numpy as np
import pandas as pd
import scanpy as sc
import seaborn as sns
import matplotlib.pyplot as plt

# ------------------------------------------------------------------
# 1️⃣  你的 celltype→Top5 基因字典
# ------------------------------------------------------------------
selection = {
    "LumSEC-Lac":  ["Fkbp11", "Slc28a3","Arhgef38", "Lalba", "Pigr"],
    "LumSEC-Lip":  ["Arhgap26", "Ltf", "Rasef",'Pdk4','Cpeb4'],
    "LumHR":       ["Ano1", "Esr1", "Gata3", "Prlr", "Stc2"],
    "Basal":       ["Acta2", "Actg2", "Myh11", "Myl9", "Tpm2"],
    "MaSC":        ["Slc1a3","Lef1", "Meis2", "Smoc2", "Snrpd1"],
}
# Derived lists preserving order
celltype_list = list(selection.keys())
gene_list     = [g for ct in celltype_list for g in selection[ct]]  # 25 genes, ordered by celltype

# ------------------------------------------------------------------
# 2️⃣  其他参数
# ------------------------------------------------------------------
datasets   = ["M-MG", "R-MG", "S-MG"]     # 样本名/文件名前缀
adata_root = Path(r"D:/111")              # 根目录
group_key  = "newcelltype"                # obs 中细胞类型列
cmap       = sns.light_palette("#404040", n_colors=256, as_cmap=True)                     # 热图配色（红-白-蓝）
dpi        = 300

sns.set(style="white", font_scale=0.8)

# ------------------------------------------------------------------
# 3️⃣  主循环
# ------------------------------------------------------------------
for ds in datasets:
    print(f"[INFO] Processing {ds} …")
    adata_path = adata_root / f"{ds}_cleaned.h5ad"
    if not adata_path.exists():
        raise FileNotFoundError(adata_path)

    adata = sc.read_h5ad(adata_path)
    if ds == "S-MG":
        if pd.api.types.is_categorical_dtype(adata.obs["newcelltype"]):
            adata.obs["newcelltype"] = adata.obs["newcelltype"].astype(str)
        adata.obs.loc[
            adata.obs["newcelltype"].isin(["MaSC", "MaSC-sg"]),
            "newcelltype"
        ] = "MaSC"
        # ③ 再转回 category，并删除未用类别
        adata.obs["newcelltype"] = (
            adata.obs["newcelltype"]
            .astype("category")
            .cat.remove_unused_categories()
        )
    adata.X=adata.layers['normalized']
    # ——— 基因 & 细胞类型实际可用列表 ————————————————
    genes_avail     = [g for g in gene_list if g in adata.var_names]
    celltypes_avail = [ct for ct in celltype_list if ct in adata.obs[group_key].unique()]

    # ——— 构建平均表达矩阵 gene × celltype ————————————
    expr = pd.DataFrame(index=genes_avail, columns=celltypes_avail, dtype=float)
    var_lookup = pd.Series(np.arange(adata.n_vars), index=adata.var_names)

    for ct in celltypes_avail:
        cells = adata.obs_names[adata.obs[group_key] == ct]
        mat   = adata[cells].X          # 假设 X 为归一化/对数化矩阵
        mean_vec = np.asarray(mat.mean(axis=0)).ravel()
        expr.loc[:, ct] = mean_vec[var_lookup[genes_avail].values]

    # ——— Z-score (行) & 旋转 ——————————
    expr_z = expr.sub(expr.mean(axis=1), axis=0)
    expr_z = expr_z.div(expr.std(axis=1).replace(0, 1), axis=0)
    heat   = expr_z.T  # 行=celltype，列=gene

    # 保证 row/col 顺序完全按 selection
    heat = heat.loc[celltypes_avail, genes_avail]

    # ——— 行颜色条 ————————————————
    if f"{group_key}_colors" in adata.uns:
        ct_color_map = dict(zip(
            adata.obs[group_key].cat.categories,
            adata.uns[f"{group_key}_colors"]
        ))
        row_colors = [ct_color_map[ct] for ct in heat.index]
    else:
        row_colors = None

    # ——— 画图 ——————————————
    figsize = (max(6, len(gene_list) * 0.35),
               max(3, len(celltypes_avail) * 0.4))

    g = sns.clustermap(
        heat,
        row_cluster=False, col_cluster=False,
        cmap=cmap,
        vmin=0, vmax=2,
        row_colors=row_colors,
        linewidths=0.1,
        figsize=figsize
    )

    g.ax_heatmap.set_xlabel("Gene")
    g.ax_heatmap.set_ylabel("Cell type")
    g.ax_heatmap.set_title(ds, pad=12)
    plt.setp(g.ax_heatmap.get_xticklabels(), rotation=90)

    out_name = f"{ds}_zscore_heatmap.pdf"
    g.savefig(out_name, dpi=dpi, bbox_inches="tight")
    plt.close()
    print(f"  [OK] Saved → {out_name}")

print("\n✅ All heatmaps generated!")
```

```python
#!/usr/bin/env python
# coding: utf-8
"""
Species-specific DEG 筛选 & Heatmap（仅绘 5 个指定细胞类型）
 - 使用 adata.layers['normalized'] 作为表达矩阵
 - 每张图行顺序固定为 CELLTYPE_ORDER
"""

from pathlib import Path
import numpy as np
import pandas as pd
import scanpy as sc
import seaborn as sns
import matplotlib.pyplot as plt

# ------------------------------------------------------------------
# ★★ 只绘制下列细胞类型，顺序固定 ★★
CELLTYPE_ORDER = ['LumSEC-Lac', 'LumSEC-Lip','LumHR','Basal','MaSC']

# 物种 → AnnData 文件名前缀
DATASETS = {"M": "M-MG", "R": "R-MG", "S": "S-MG"}

# 对应 DEG csv 路径
DEG_PATH = {
    sp: Path(f"D:/111/{name}_ranked_genes.csv")
    for sp, name in DATASETS.items()
}

ADATA_ROOT = Path(r"D:/111")
GROUP_KEY  = "newcelltype"
ZERO_EXPR_THRESHOLD = 0.01
TOP_N      = 5
HEAT_CMAP  = "flare_r"
DPI        = 300

# ------------------------------------------------------------------
SPECIES_CMAP = {
    "M": sns.light_palette("#E5B764", n_colors=256, as_cmap=True),   # Mouse
    "R": sns.light_palette("#7BC6E6", n_colors=256, as_cmap=True),   # Rabbit
    "S": sns.light_palette("#A3C29F", n_colors=256, as_cmap=True),   # Sugar-glider
}
# ------------------------------------------------------------------


def load_all_anndata():
    """读 3 份 AnnData 到 dict{species: AnnData}；确保有 'normalized' layer."""
    ads = {}
    for sp, tag in DATASETS.items():
        f = ADATA_ROOT / f"{tag}_cleaned.h5ad"
        ad = sc.read_h5ad(f)
        if tag == "S-MG":
            if pd.api.types.is_categorical_dtype(ad.obs["newcelltype"]):
                ad.obs["newcelltype"] = ad.obs["newcelltype"].astype(str)
            ad.obs.loc[
                ad.obs["newcelltype"].isin(["MaSC", "MaSC-sg"]),
                "newcelltype"
            ] = "MaSC"
            ad.obs["newcelltype"] = (
                ad.obs["newcelltype"]
                .astype("category")
                .cat.remove_unused_categories()
            )
        if 'normalized' not in ad.layers:
            raise KeyError(f"{f} 缺少 layers['normalized']")
        ads[sp] = ad
    return ads


def genes_zero_in_others(genes, other_adatas):
    """过滤：在所有“其他物种”里 ≤5% 细胞表达的基因集合."""
    keep = set(genes)
    for ad in other_adatas:
        mat_other = (ad[:, list(keep)].layers['normalized'] > 0)
        prop = mat_other.mean(axis=0).A1
        keep &= {g for g, p in zip(keep, prop) if p <= ZERO_EXPR_THRESHOLD}
    return keep


def filter_species_specific(sp, ads):
    """返回 dict{celltype: DEG-DataFrame}，仅保留 CELLTYPE_ORDER 内的 celltype."""
    ad_target = ads[sp]
    others    = [ads[o] for o in ads if o != sp]

    zero_in_others = genes_zero_in_others(ad_target.var_names, others)

    deg = pd.read_csv(DEG_PATH[sp])
    deg = deg[deg["gene"].isin(zero_in_others)]
    deg = deg.sort_values("logfoldchange", ascending=False)

    out = {ct: df for ct, df in deg.groupby("group") if ct in CELLTYPE_ORDER}
    return out


def save_deg_csv(ct_dict, sp):
    df = pd.concat(
        [d.assign(celltype=ct) for ct, d in ct_dict.items()],
        ignore_index=True
    )
    out = f"{sp}_specific_DEG.csv"
    df.to_csv(out, index=False)
    return out


def top_genes(ct_dict, n=TOP_N):
    genes = []
    for ct in CELLTYPE_ORDER:
        if ct in ct_dict:
            genes += ct_dict[ct].nlargest(n, 'logfoldchange')['gene'].tolist()
    # 去重但保持第一次出现的顺序
    seen = set(); ordered = []
    for g in genes:
        if g not in seen:
            ordered.append(g); seen.add(g)
    return ordered


def plot_heatmap(genes, adata, dataset_tag, species_tag):
    """仅绘 CELLTYPE_ORDER 行；表达矩阵来自 layers['normalized']."""
    present_ct = [ct for ct in CELLTYPE_ORDER if ct in adata.obs[GROUP_KEY].unique()]
    expr = pd.DataFrame(index=genes, columns=present_ct, dtype=float)
    var_lookup = pd.Series(np.arange(adata.n_vars), index=adata.var_names)

    for ct in present_ct:
        cells = adata.obs_names[adata.obs[GROUP_KEY] == ct]
        mat   = adata[cells].layers['normalized']
        mean_vec = np.asarray(mat.mean(axis=0)).ravel()
        expr[ct] = mean_vec[var_lookup[genes].values]

    # Z-score (每基因行) 并截断到 [0,2]
    expr_z = (
        expr.sub(expr.mean(axis=1), axis=0)
            .div(expr.std(axis=1).replace(0, 1), axis=0)
            .clip(lower=0, upper=2)      # 负值→0；>2→2
            .T
    ).loc[present_ct]                    # 保证行顺序

    # 行颜色（若 AnnData.uns 提供调色板）
    row_colors = None
    if f"{GROUP_KEY}_colors" in adata.uns:
        cat      = adata.obs[GROUP_KEY].cat.categories
        palette  = adata.uns[f"{GROUP_KEY}_colors"]
        lut      = dict(zip(cat, palette))
        row_colors = [lut[ct] for ct in present_ct]

    # --- 选物种专属色板 ---
    cmap_choice = SPECIES_CMAP.get(species_tag, HEAT_CMAP)

    sns.set(style="white", font_scale=0.75)
    g = sns.clustermap(
        expr_z,
        row_cluster=False, col_cluster=False,
        cmap=cmap_choice,
        vmin=0, vmax=2,              # 固定色阶 0→浅, 2→深
        row_colors=row_colors,
        linewidths=0.1,
        figsize=(max(6, len(genes)*0.4), max(3, len(present_ct)*0.45))
    )
    g.ax_heatmap.set_xlabel("Gene")
    g.ax_heatmap.set_ylabel("Cell type")
    g.ax_heatmap.set_title(f"{species_tag} TopGenes on {dataset_tag}", pad=10)
    plt.setp(g.ax_heatmap.get_xticklabels(), rotation=90)

    fname = f"{species_tag}_TopGenes_on_{dataset_tag}.pdf"
    g.savefig(fname, dpi=DPI, bbox_inches="tight")
    plt.close()
    return fname


# =============================== MAIN ===============================
def main():
    ads = load_all_anndata()
    top_dict = {}

    for sp in DATASETS:
        print(f"\n[ {sp} ] 筛选物种特异基因 …")
        ct_deg = filter_species_specific(sp, ads)
        csvfile = save_deg_csv(ct_deg, sp)
        print(f"  CSV → {csvfile}")
        top_dict[sp] = top_genes(ct_deg)
        print(f"  Top genes ({len(top_dict[sp])}) collected.")

    # 绘 9 张图
    for sp, genes in top_dict.items():
        for ds_sp, ds_tag in DATASETS.items():
            fname = plot_heatmap(genes, ads[ds_sp], ds_tag, sp)
            print(f"  Heatmap saved: {fname}")

    print("\n✅ 任务完成！共生成 3 个 CSV、9 张 heatmap。")

if __name__ == "__main__":
    main()
```

```python
# ---------------------------------------------------------------
# Imports & constants
# ---------------------------------------------------------------
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from pathlib import Path

data_dir = Path(".")          # ← 若 CSV 不在当前目录，请改这里

CELLTYPE_ORDER = ["LumSEC-Lac", "LumSEC-Lip", "LumHR", "Basal", "MaSC"]
DATASETS       = {"M": "M-MG", "R": "R-MG", "S": "S-MG"}

label_colors = {
    "Basal":        "#1f78b4",
    "LumHR":        "#ff7f00",
    "LumSEC-Lac":   "#6a3d9a",
    "LumSEC-Lip":   "#b15928",
    "MaSC":         "#33a02c",
}
hatches = {"M": "", "R": "//", "S": "++"}

PDF_DPI        = 300
CONS_BAR_H     = 0.6                 # 厚度（右图单条）
SPEC_BAR_H     = CONS_BAR_H / 3      # 左图单条 = 右图总厚度 / 3
# ---------------------------------------------------------------
# 1) 统计物种特异 (logFC>0)
# ---------------------------------------------------------------
spec_counts = {}
for sp in DATASETS:
    f = data_dir / f"{sp}_specific_DEG.csv"
    df = pd.read_csv(f)
    df = df[df["logfoldchange"] > 0]
    counts = (df.groupby("celltype")
                .size()
                .reindex(CELLTYPE_ORDER)
                .fillna(0)
                .astype(int))
    spec_counts[sp] = counts
spec_df = pd.DataFrame(spec_counts)
# ---------------------------------------------------------------
# 2) 统计保守基因
# ---------------------------------------------------------------
cons_f = data_dir / "conserved_gene_expression.csv"
cons_df = pd.read_csv(cons_f)
cons_counts = (cons_df.groupby("celltype")
                         .size()
                         .reindex(CELLTYPE_ORDER)
                         .fillna(0)
                         .astype(int))
# ---------------------------------------------------------------
# 3) 绘图
# ---------------------------------------------------------------
y_pos = range(len(CELLTYPE_ORDER))
fig, (axL, axR) = plt.subplots(
    1, 2, figsize=(12, 4.5),
    sharey=True,
    gridspec_kw={"width_ratios": [2, 1]}
)

# ---- 左：物种特异（镜像、合适厚度） ----
for i, sp in enumerate(DATASETS):
    offset = (i - 1) * SPEC_BAR_H          # -SPEC, 0, +SPEC
    axL.barh(
        [y + offset for y in y_pos],
        -spec_df[sp],                      # 负值 → 向左
        height=SPEC_BAR_H,
        color=[label_colors[ct] for ct in CELLTYPE_ORDER],
        hatch=hatches[sp],
        edgecolor="black", linewidth=0.6
    )

# 去掉左图 y 轴标签和左侧 spine
axL.set_yticks([])
axL.set_ylabel("")
axL.spines["left"].set_visible(False)

axL.set_xlabel("number of markers")
axL.set_title("Species-specific (logFC > 0)")
axL.invert_yaxis()

# 将负刻度标签显示为正数
axL.set_xticklabels([abs(int(x)) for x in axL.get_xticks()])

# 图例：空心 patch
legend_handles = [
    mpatches.Patch(facecolor="none",
                   hatch=hatches[sp],
                   edgecolor="black",
                   label=f"{sp} ({DATASETS[sp]})")
    for sp in DATASETS
]
axL.legend(handles=legend_handles,
           frameon=False, loc="upper left")

# ---- 右：保守 ----
axR.barh(
    y_pos,
    cons_counts,
    height=CONS_BAR_H,
    color=[label_colors[ct] for ct in CELLTYPE_ORDER],
    edgecolor="black", linewidth=0.6
)
axR.set_xlabel("number of markers")
axR.set_title("Conserved")
axR.invert_yaxis()

# ---- 坐标轴 & 边框处理 ----
for ax in (axL, axR):
    for spine in ("top", "right"):
        ax.spines[spine].set_visible(False)
    ax.spines["bottom"].set_linewidth(0.8)

# 右图保留 ytick 标签
axR.set_yticks(y_pos)
axR.set_yticklabels(CELLTYPE_ORDER)

plt.tight_layout()
plt.savefig("marker_counts_combined.pdf", dpi=PDF_DPI)
plt.close()

print("✓ marker_counts_combined.pdf 已生成")
```
